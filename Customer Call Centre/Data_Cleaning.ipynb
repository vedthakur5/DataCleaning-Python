```python
import pandas as pd
df = pd.read_excel("Customer Call List.xlsx")
df
```

```python
df.drop_duplicates()
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop(columns = "Not_Useful_Column")
```

```python
# df["Last_Name"] = df["Last_Name"].str.lstrip("...")
# df["Last_Name"] = df["Last_Name"].str.lstrip("/")
# df["Last_Name"] = df["Last_Name"].str.rstrip("_")
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].replace(r'\W', '', regex=True)
```

```python
df['Phone_Number'] = df['Phone_Number'].astype(str)
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('nan--', '')
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('Na--', '')
```

```python
df[['Street_Name', 'State', 'Zip_Code']] = df['Address'].str.split(',', expand=True)
```

```python
df = df.drop(columns = "Address")
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].replace('Y', 'Yes')
```

```python
df['Paying Customer'] = df['Paying Customer'].replace('N', 'No')
```

```python
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('Y', 'Yes')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('N', 'No')
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].replace('N/a', '')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('NaN', '')
```

```python
df = df.fillna('')
```

```python
df
```

```python
df = df.drop(df[df['Do_Not_Contact'] == 'Yes'].index)
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('', 'No')
```

```python
df = df.drop(df[df['Phone_Number'] == ''].index)
```

```python
df
```

```python
df = df.reset_index(drop = True)
```
